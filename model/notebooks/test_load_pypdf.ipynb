```python
from langchain.document_loaders import PyPDFLoader
from langchain_openai import ChatOpenAI
import os
import pathlib
```

```python
llm = ChatOpenAI(model="gpt-3.5-turbo-1106", temperature=0)
```

```python
def iterate_pdfs(dir='references'):
    for path, subdirs, files in os.walk(dir):
        for name in files:
            if name.endswith('.pdf'):
                fp=pathlib.PurePath(path, name)
                yield PyPDFLoader(fp.as_posix()).load()
```

```python
pdfer = iterate_pdfs()
```

```python
doc= pdfer.__next__()
```
